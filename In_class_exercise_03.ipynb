```python
# You code here (Please add comments in the code):

import pandas as pd
from bs4 import BeautifulSoup
import urllib
import requests
```

```python
page = urllib.request.urlopen(urllib.request.Request("https://www.worldometers.info/coronavirus/",None,{'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7',}))

soup = BeautifulSoup(page)
# print(soup.prettify)
```

```python
data_table = soup.find_all("tr",{"style":""})
number = soup.find_all("td",{"style":"font-size:12px;color: grey;text-align:center;vertical-align:middle;"})
country = soup.find_all("a",{"class":"mt_a"})
# total_cases = soup.find_all("td",{"style":"font-weight: bold; font-size:15px; text-align:left;"})
newCases = soup.find_all("td",{"style":"font-weight: bold; text-align:right;background-color:#FFEEAA;"})
totalDeaths = soup.find_all("td",{"style":"font-weight: bold; text-align:right;"})
newDeaths = soup.find_all("td",{"style":"font-weight: bold; \n                                    text-align:right;background-color:red; color:white"})
list_of_values = soup.find_all("td",{"style":"font-weight: bold; text-align:right"})
#print(list_of_values)
activeCases = soup.find_all("td",{"style":"text-align:right;font-weight:bold;"})
values = list()
```

```python
#-------------------------------------------------

numbers_l = list()
country_names =list()
total_recovered = list()
serious_or_critical = list()
tot_cases_1mpop = list()
deaths_1Mpop = list()
total_tests = list()
tests_1Mpop = list()
population=list()
total_cases = list()
new_cases = list()
total_deaths = list()
new_deaths =list()
active_cases = list()
for i in list_of_values:
  values.append(i.text)

for i in range(0,len(values),8):
  total_cases.append(values[i])
  total_recovered.append(values[i+1])
  serious_or_critical.append(values[i+2])
  tot_cases_1mpop.append(values[i+3])
  deaths_1Mpop.append(values[i+4])
  total_tests.append(values[i+5])
  tests_1Mpop.append(values[i+6])
  population.append(values[i+7])


for i,c,n,t,nd,ac in zip(number,country,newCases,totalDeaths,newDeaths,activeCases):
  numbers_l.append(i.text)
  country_names.append(c.text)
  new_cases.append(n.text)
  total_deaths.append(t.text)
  new_deaths.append(nd.text)
  active_cases.append(ac.text)

ouputDF = pd.DataFrame(list(zip(country_names,total_cases,new_cases,total_deaths,new_deaths,total_recovered,active_cases,serious_or_critical,tot_cases_1mpop,deaths_1Mpop,total_tests,tests_1Mpop,population)),\
                       columns = ['country','total_cases','new_cases','total_deaths','new_deaths','total_recovered','active_cases','serious_or_critical','total_cases_1m_Population','Deaths_For_1M_population','Total_Tests','Tests_for_1M_Population','Population'])

pd.set_option("display.max_rows", 500, "display.max_columns", 50) 
ouputDF
```
