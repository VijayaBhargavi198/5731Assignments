(2) Write a Python program to count the number of characters in a string. (2 points)

Sample String : 'google.com‘, 

Expected Result : {'o': 3, 'g': 2, '.': 1, 'e': 1, 'l': 1, 'm': 1, 'c': 1}

```python
def character_count(string):
    dictionary = {}
    for n in string:
        keys = dictionary.keys()
        if n in keys:
            dictionary[n] += 1
        else:
            dictionary[n] =1   
    return dictionary
print(character_count("apple"))
        
```

(4) Write a Python program to get a string from a given string where all occurrences of its first char have been changed to '\$', except the first char itself. (2 points)

Sample String : 'restart‘, 

Expected Result : 'resta\$t'

```python
def repeating_letter(string1):
    character = string1[0]
    string1 = string1.replace(character, '$')
    string1 = character + string1[1:]
    return string1

print(repeating_letter('bubble'))
```

(6) Write a Python program to concatenate following dictionaries to create a new one. (3 points)

Sample Dictionary : dic1={1:10, 2:20}, dic2={3:30, 4:40}, dic3={5:50,6:60}, 

Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}


```python
def concatination(dict1, dict2, dict3): 
    result = {**dict1, **dict2, **dict3} 
    return result
dict1={1:10, 2:20,7:70}
dict2={3:30, 4:40,8:80}
dict3={5:50,6:60,9:90} 
resulting_dict = concatination(dict1, dict2, dict3) 
print(resulting_dict)
```
